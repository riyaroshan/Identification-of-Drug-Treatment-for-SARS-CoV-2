```python
#generating images from the pixel values

from PIL import Image
import numpy as np

def generate_images(df,name):
    width = 32
    height = 32

    for i, row in df.iterrows():
        pixels = np.array(row).reshape((height, width))
        
        image = Image.fromarray(np.uint8(pixels))
        filename = f"image_{i}.png"
        image.save(f'dataset/train/{name}/{filename}')
        print(f"Saved image {i}")

generate_images(df1,'remdesivir')
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Define the paths to the original image folder and the new train/validation 
original_dir = os.getcwd()+'\\dataset\\overall'
train_dir = os.getcwd()+'\\dataset\\train'
val_dir = os.getcwd()+'\\dataset\\valid'

# Define the ratio of validation set
val_ratio = 0.3

# Split the images into train and validation sets
classes = os.listdir(original_dir)
for cls in classes:
    cls_path = os.path.join(original_dir, cls)
    images = os.listdir(cls_path)
    train_images, val_images = train_test_split(images, test_size=val_ratio)
    
    # Move the train images to the train folder
    train_cls_path = os.path.join(train_dir, cls)
    os.makedirs(train_cls_path, exist_ok=True)  
    for image in train_images:
        src_path = os.path.join(cls_path, image)
        dst_path = os.path.join(train_cls_path, image)
        shutil.copy(src_path, dst_path)
    
    # Move the validation images to the validation folder
    val_cls_path = os.path.join(val_dir, cls)
    os.makedirs(val_cls_path, exist_ok=True)
    for image in val_images:
        src_path = os.path.join(cls_path, image)
        dst_path = os.path.join(val_cls_path, image)
        shutil.copy(src_path, dst_path)
```

```python
import splitfolders
import tensorflow as tf
from keras import layers, models
from keras.preprocessing.image import ImageDataGenerator

# Split dataset into train and validation sets using split-folders library
input_folder = os.getcwd()+'\\dataset\\overall'
output_folder = os.getcwd()+'\\output'
splitfolders.ratio(input_folder, output_folder, seed=42, ratio=(.8, .2))
```

```python
# Define image data generators for training and validation sets
train_datagen = ImageDataGenerator(rescale=1./255)
valid_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    output_folder+'/train',
    target_size=(32, 32),
    color_mode="grayscale",
    batch_size=32,
    class_mode='categorical')

valid_generator = valid_datagen.flow_from_directory(
    output_folder+'/val',
    target_size=(32, 32),
    color_mode="grayscale",
    batch_size=32,
    class_mode='categorical')

# Define CNN model architecture
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(5, activation='softmax')
])

# Compile model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train model
model.fit(train_generator, epochs=10, validation_data=valid_generator)
```
